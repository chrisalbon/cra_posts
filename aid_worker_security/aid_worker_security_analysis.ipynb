```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt

# Set ipython's max row display
pd.set_option('display.max_row', 1000)

# Set iPython's max column width to 50
pd.set_option('display.max_columns', 50)

# Set iPython to display graphics inline
%matplotlib inline
```

```python
# Read in the csv file using latin-1 encoding
df = pd.read_csv('data/data_2.csv', encoding='latin-1')
```

```python
# View the first three lines of the dataset
df.head(3)
```

```python
# Create a new dataframe that groups observations by year
# and summarizes numeric cells
by_year = df.groupby('Year').sum()
```

```python
# View the first three rows of that dataframe
by_year.head(3)
```

```python
# Drop the last year (because the data is incomplete)
by_year = by_year.drop(2014)
```

```python
# Plot the numbers of victims each year and in total
by_year[['Total affected', 'Total killed', 'Total wounded', 'Total kidnapped']].plot(grid=False)
```

```python
# Create a list of the mean scores for each variable
sum_values = [df['Internationals killed'].sum(), df['Internationals wounded'].sum(), df['Internationals kidnapped'].sum()]

# Create a list of variances, which are set at .25 above and below the score
variance = [df['Internationals killed'].sum() * 0.25, df['Internationals killed'].sum() * 0.25, df['Internationals killed'].sum() * 0.25]

# Set the bar labels
bar_labels = ['Killed', 'Wounded', 'Kidnapped']

# Create the x position of the bars
x_pos = list(range(len(bar_labels)))

# Create the plot bars
# In x position
plt.bar(x_pos,
        # using the data from the mean_values
        sum_values, 
        # aligned in the center
        align='center',
        # with color
        color='#FFC222',
        # alpha 0.5
        alpha=1)

# add a grid
plt.grid()

# set height of the y-axis
max_y = max(zip(sum_values, variance)) # returns a tuple, here: (3, 5)
plt.ylim([0, (max_y[0] + max_y[1]) * 1.1])

# set axes labels and title
plt.ylabel('Number Affected')
plt.xticks(x_pos, bar_labels)
plt.title('Total International Aid Workers Affected By Type (1996-2014)')

plt.show()
```

```python
# Create a list of the mean scores for each variable
sum_values = [df['Nationals killed'].sum(), df['Nationals wounded'].sum(), df['Nationals kidnapped'].sum()]

# Create a list of variances, which are set at .25 above and below the score
variance = [df['Nationals killed'].sum() * 0.25, df['Nationals killed'].sum() * 0.25, df['Nationals killed'].sum() * 0.25]

# Set the bar labels
bar_labels = ['Killed', 'Wounded', 'Kidnapped']

# Create the x position of the bars
x_pos = list(range(len(bar_labels)))

# Create the plot bars
# In x position
plt.bar(x_pos,
        # using the data from the mean_values
        sum_values, 
        # aligned in the center
        align='center',
        # with color
        color='#FFC222',
        # alpha 0.5
        alpha=1)

# add a grid
plt.grid()

# set height of the y-axis
max_y = max(zip(sum_values, variance)) # returns a tuple, here: (3, 5)
plt.ylim([0, (max_y[0] + max_y[1]) * 1.1])

# set axes labels and title
plt.ylabel('Number Affected')
plt.xticks(x_pos, bar_labels)
plt.title('Total Nationals Affected By Type (1996-2014)')

plt.show()
```

```python
# Create a figure with a single subplot
f, ax = plt.subplots(1, figsize=(10,5))

# Set bar width at 1
bar_width = 1

# positions of the left bar-boundaries
bar_l = [i for i in range(len(by_year['Nationals killed']))] 

# positions of the x-axis ticks (center of the bars as bar labels)
tick_pos = [i+(bar_width/2) for i in bar_l] 

# Create the total score for each participant
totals = [i+j+k for i,j,k in zip(by_year['Nationals killed'], 
                                 by_year['Nationals wounded'], 
                                 by_year['Nationals kidnapped'])]

# Create the percentage of the total score the pre_score value for each participant was
pre_rel = [i / j * 100 for  i,j in zip(by_year['Nationals killed'], totals)]

# Create the percentage of the total score the mid_score value for each participant was
mid_rel = [i / j * 100 for  i,j in zip(by_year['Nationals wounded'], totals)]

# Create the percentage of the total score the post_score value for each participant was
post_rel = [i / j * 100 for  i,j in zip(by_year['Nationals kidnapped'], totals)]

# Create a bar chart in position bar_1
ax.bar(bar_l, 
       # using pre_rel data
       pre_rel, 
       # labeled 
       label='Nationals killed', 
       # with alpha
       alpha=0.9, 
       # with color
       color='#019600',
       # with bar width
       width=bar_width,
       # with border color
       edgecolor='white'
       )

# Create a bar chart in position bar_1
ax.bar(bar_l, 
       # using mid_rel data
       mid_rel, 
       # with pre_rel
       bottom=pre_rel, 
       # labeled 
       label='Nationals wounded', 
       # with alpha
       alpha=0.9, 
       # with color
       color='#3C5F5A', 
       # with bar width
       width=bar_width,
       # with border color
       edgecolor='white'
       )

# Create a bar chart in position bar_1
ax.bar(bar_l, 
       # using post_rel data
       post_rel, 
       # with pre_rel and mid_rel on bottom
       bottom=[i+j for i,j in zip(pre_rel, mid_rel)], 
       # labeled 
       label='Nationals kidnapped',
       # with alpha
       alpha=0.9, 
       # with color
       color='#219AD8', 
       # with bar width
       width=bar_width,
       # with border color
       edgecolor='white'
       )

# Set the ticks to be first names
plt.xticks(tick_pos, by_year.index)
ax.set_ylabel("Proportion Of Attack Type")
ax.set_xlabel("Year")

# Let the borders of the graphic
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
plt.ylim(-10, 110)

# rotate axis labels
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

# shot plot
plt.show()
```
