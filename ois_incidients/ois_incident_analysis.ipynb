```python
import requests
import pandas as pd
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Set the max number of row and columns displayed,
# as well as the number of characters displayed in each cell
pd.set_option('display.max_row', 1000)
pd.set_option('display.max_column', 50)
pd.set_option('max_colwidth', 500)
```

```python
sns.set_style('whitegrid')
```

```python
url = 'http://ois-incidents.appspot.com/content?limit=100000'

r = requests.get(url)
df = pd.DataFrame(r.json())
```

```python
df = df['results'].apply(pd.Series)
```

```python
outcome_count = df.groupby(['outcome']).count()

outcome_count = outcome_count.drop(['agency','city','county','incident_date',
                                'officer_names','victim_armed','searched_date',
                                'source_url','state','summary','victim_age',
                                'victim_race','victim_gender',
                                'victim_name', 'shots_fired', 'weapon'], axis=1)

outcome_count = outcome_count.rename(columns={"shootings": "count"})
outcome_count = outcome_count.sort(columns='count', ascending=False)

outcome_count['percent'] = outcome_count['count'] / outcome_count['count'].sum() * 100
outcome_count['percent'] = outcome_count['percent'].round(decimals=1)
```

```python
sns.set_context("talk", font_scale=1.1)


sns.barplot(outcome_count.index,
            'percent',
            color='#ef8a62',
            data=outcome_count)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
sns.despine(left=True)
plt.title('Outcome of Officer Involved Incidents')
plt.xlabel('Incident Outcome')
plt.ylabel('Percent of Incidents')
```

```python
armed_count = df.groupby(['victim_armed']).count()

armed_count = armed_count.drop(['agency','city','county','incident_date',
                                'officer_names','outcome','searched_date',
                                'source_url','state','summary','victim_age',
                                'victim_race','victim_gender',
                                'victim_name', 'shots_fired', 'weapon'], axis=1)

armed_count = armed_count.rename(columns={"shootings": "count"})
armed_count = armed_count.sort(columns='count', ascending=False)
```

```python
sns.set_context("talk", font_scale=1.1)

sns.barplot(armed_count.index,
            'count',
            color='#ef8a62',
            data=armed_count)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
sns.despine(left=True)
plt.title('Number of Incidents by Race')
plt.xlabel('Race of Victim')
plt.ylabel('Number of Incidents')
```

```python
race_count = df.groupby(['victim_race']).count()

race_count = race_count.drop(['agency','city','county','incident_date',
                                'officer_names','outcome','searched_date',
                                'source_url','state','summary','victim_age',
                                'victim_armed','victim_gender',
                                'victim_name', 'shots_fired', 'weapon'], axis=1)

race_count = race_count.rename(columns={"shootings": "count"})
race_count = race_count.sort(columns='count', ascending=False)
```

```python
sns.set_context("talk", font_scale=1.1)

sns.barplot(race_count.index,
            'count',
            color='#ef8a62',
            x_order=race_count.index,
            data=race_count)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
sns.despine(left=True)
plt.title('Number of Incidents by Race')
plt.xlabel('Race of Victim')
plt.ylabel('Number of Incidents')
```

```python
gender_count = df.groupby(['victim_gender']).count()

gender_count = gender_count.drop(['agency','city','county','incident_date',
                                'officer_names','outcome','searched_date',
                                'source_url','state','summary','victim_age',
                                'victim_armed','victim_race',
                                'victim_name', 'shots_fired', 'weapon'], axis=1)

gender_count = gender_count.rename(columns={"shootings": "count"})
gender_count['gender'] = gender_count.index
gender_count = gender_count.sort(columns='count', ascending=False)
```

```python
sns.set_context("talk", font_scale=1.1)

sns.barplot('gender',
            'count',
            color='#ef8a62',
            x_order=gender_count.index,
            data=gender_count)
locs, labels = plt.xticks()
sns.despine(left=True)
plt.title('Number of Incidents by Gender')
plt.xlabel('Gender of Victim')
plt.ylabel('Number of Incidents')
```

```python
shots_count = df.groupby(['shots_fired']).count()

shots_count = shots_count.drop(['agency','city','county','incident_date',
                                'officer_names','outcome','searched_date',
                                'source_url','state','summary','victim_age',
                                'victim_armed','victim_gender','victim_race',
                                'victim_name','weapon'], axis=1)

shots_count = shots_count.rename(columns={"shootings": "count"})

shots_count['shots'] = shots_count.index
shots_count['shots'] = shots_count['shots'].astype(int)

shots_count['percent'] = shots_count['count'] / shots_count['count'].sum() * 100
shots_count['percent'] = shots_count['percent'].round(decimals=1)
```

```python
sns.set_context("talk", font_scale=1.1)

sns.barplot('shots',
            'count',
            color='#ef8a62',
            data=shots_count)
locs, labels = plt.xticks()
sns.despine(left=True)
plt.setp(labels, rotation=90)
plt.title('Number Of Incidents Per Shots Fired')
plt.xlabel('Shots Fired')
plt.ylabel('Number of Incidents')
```

```python
sns.set_context("talk", font_scale=1.1)

sns.barplot('shots',
            'percent',
            color='#ef8a62',
            data=shots_count)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
sns.despine(left=True)
plt.title('Incident Percent Per Shots Fired')
plt.xlabel('Shots Fired')
plt.ylabel('Percent of Incidents')
```

```python
age_count = df.groupby(['victim_age']).count()

age_count = age_count.drop(['agency','city','county','incident_date',
                                'officer_names','outcome','searched_date',
                                'source_url','state','summary','victim_armed',
                                'victim_race','victim_gender',
                                'victim_name', 'shots_fired', 'weapon'], axis=1)

age_count = age_count.rename(columns={"shootings": "count"})
age_count['victim_age'] = age_count.index
```

```python
age_count.head()
```

```python
years = pd.DataFrame()
years['all_ages'] = range(0,100, 1)
```

```python
years = pd.merge(years, age_count, left_on='all_ages', right_on='victim_age', how='left')
```

```python
sns.set_context("talk", font_scale=1.1)

sns.barplot('all_ages',
            'count',
            color='#ef8a62',
            data=years)
sns.despine(left=True)
plt.xticks(range(0, 100, 10), [str(x) for x in range(0, 100, 10)])  
plt.title('Number Of Incidents By Age of Victim')
plt.xlabel('Age of Victim')
plt.ylabel('Number of Incidents')
```
